```python
import json 

full_endo_json = 'endoscopy/full_endo.json'

full_endo_data = json.load(open(full_endo_json))

final_object = {}
position = ['stomach','colorectal polyps','esophagus','doudenum']
color = ['WLI','LCI','FICE','BLI']
HP_cond = ['positive HP','negative HP']
id = 0

for entry in full_endo_data:
    id +=1
    data_dict = {
        'image': entry.split('/')[-1],
        'mask': entry.split('/')[-1].split('.')[0] + '.png'
    }
    for key in full_endo_data[entry]:
        if key in position:
            if key == 'colorectal polyps':
                data_dict['position'] = 'colon'
            elif key == 'doudenum':
                data_dict['position'] = 'duodenum'
            else:
                data_dict['position'] = key
        elif key in color:
            data_dict['color'] = key
        elif key in HP_cond:
            data_dict['HP'] = key[:8]
        else:
            if data_dict.get('lesion') is None:
                data_dict['lesion'] = []
            data_dict['lesion'].append(key)
        
    final_object[id] = data_dict

final_object = json.dumps(final_object, indent=4)
with open('endoscopy/metadata.json', 'w') as file:
    file.write(final_object)
```

```python
import random as rd
max = 31939

# split in to 90% train and 10% test
train = rd.sample(range(1,max+1), int(max*0.9))
test = list(set(range(1,max+1)) - set(train))

with open('endoscopy/train.txt', 'w') as file:
    for i in train:
        file.write(str(i) + '\n')
    file.close()
with open('endoscopy/test.txt', 'w') as file:
    for i in test:
        file.write(str(i) + '\n')
    file.close()

# train = rd.sample(range(1,max+1), int(max*0.8))
# val_test = list(set(range(1,max+1)) - set(train))
# val = rd.sample(set(val_test), int(val_test * 0.5))
# test = list(set(set(val_test) - set(val)))

# with open('endoscopy/train.txt', 'w') as file:
#     for i in train:
#         file.write(str(i) + '\n')
#     file.close()

# with open('endoscopy/val.txt', 'w') as file:
#     for i in val:
#         file.write(str(i) + '\n')
#     file.close()

# with open('endoscopy/test.txt', 'w') as file:
#     for i in test:
#         file.write(str(i) + '\n')
#     file.close()

```
